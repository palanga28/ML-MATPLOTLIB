## 1. Importer les librairies

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

## 2.Charger le dataset depuis UCI



```python
heart_disease = fetch_ucirepo(id=45)
```

```python
data = heart_disease.data.features.copy()
```

## 3.1 visualisation de donnes avec matplotlib

```python
plt.figure()
plt.hist(data['age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Nombre de patients')
plt.title('Distribution de l’âge')
plt.show()
```

```python
data.hist(figsize=(15, 10))
plt.suptitle("Distribution de toutes les variables", fontsize=16)
plt.show()
```

```python
plt.figure()
data['sex'].value_counts().plot(kind='bar')
plt.xlabel('Sexe (0=femme, 1=homme)')
plt.ylabel('Nombre')
plt.title('Répartition par sexe')
plt.show()
```
